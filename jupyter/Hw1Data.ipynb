```python
import glob
import hashlib

import pandas as pd
import numpy as np
```

```python
def md5_hash(user: int, prefix="9348yhfs9hrwurnsdmu4") -> str:
    encoded_text = f"{prefix}{user}".encode('utf-8')
    md5_object = hashlib.md5(encoded_text)
    return md5_object.hexdigest()
```

```python
train_data = pd.concat([
    pd.read_json(data_path, lines=True)
    for data_path
    in glob.glob("/Users/n.anokhin/Desktop/train/*/data.json")
])


train_data["user"] = train_data["user"].map(md5_hash)

train_data[["user", "track", "time"]].to_csv("/Users/n.anokhin/Desktop/train.csv", index=False)
```

```python
test_data = pd.concat([
    pd.read_json(data_path, lines=True)
    for data_path
    in glob.glob("/Users/n.anokhin/Desktop/test/*/data.json")
])

test_data["user"] = test_data["user"].map(md5_hash)
test_data["public"] = test_data["user"].map(lambda u: hash(u) % 2 == 0)
test_data["score"] = np.random.random(size=len(test_data))

test_data[["user", "track", "time", "public"]].to_csv("/Users/n.anokhin/Desktop/test_truth.csv", index=False)
test_data[["user", "track"]].to_csv("/Users/n.anokhin/Desktop/test.csv", index=False)
test_data[["user", "track", "score"]].to_csv("/Users/n.anokhin/Desktop/submit.csv", index=False)
```
